```python
import os
```

```python
full_dirs = []
for dir_path, path_names, filenames in os.walk('F:\Gitclone_file'):
    for filename in filenames:
        full_dir = os.path.join(dir_path, filename)
        full_dirs.append(full_dir)
print(len(full_dirs))
```

```python
NEW_LINECHAR = '<N>'
MIN_STR_LEN = 256
MAX_STR_LEN = 512
count = 0
with open('smaller.txt', 'a') as f: 
    for full_dir in full_dirs:
        try:
            print(full_dir)
            fd = open(full_dir, 'r').read()
            fd = fd.replace('\n', NEW_LINECHAR)

            if MIN_STR_LEN < len(fd) < MAX_STR_LEN:
                f.write(fd + '\n')
            else:
                substring = ''
                fd_split = fd.split(f'{NEW_LINECHAR}')
                for split in fd_split:
                    substring += split + f'{NEW_LINECHAR}{NEW_LINECHAR}'
                    if MIN_STR_LEN < len(substring) < MAX_STR_LEN:
                        f.write(substring+'\n')
                        substring = ''


            # print(fd)
            # print(len(fd))
        
        except Exception as e:
            print(str(e))

        # print()

        count += 1
        if count > 100:
            break
```
